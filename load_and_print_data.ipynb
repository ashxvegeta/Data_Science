```python
import json
```

```python
# load the data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("storedata.json")
```

```python
# clean and structured data
def clean_data(data):
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}

    for user in data:
        # clean rating
        raw_rating = user["rating"].strip().lower()
        # check if raw rating is in text to num map, if it is, convert the rating into int
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        # print(raw_rating)
        # override the user rating with the correct format
        user["rating"] = raw_rating
        # handle missing data
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None
            
        print(raw_age)
```

```python
clean_data(data)
```
