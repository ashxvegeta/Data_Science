```python
import json
```

```python
# load the data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("storedata.json")
```

```python
# clean and structured data
def clean_data(data):
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}
    clean_data = []
    unique_users = set()

    for user in data:
        # clean rating
        raw_rating = str(user["rating"]).strip().lower()
        # check if raw rating is in text to num map, if it is, convert the rating into int
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        # print(raw_rating)
        # override the user rating with the correct format
        user["rating"] = raw_rating
        # handle missing data
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None
            
        if (user["name"].strip() in unique_users):
            continue   
        unique_users.add(user["name"])
        clean_data.append(user)
        
    return clean_data
```

```python
data = clean_data(data)
```

```python
# get meaningful insight
def get_insight(data):
    # avg rating
    total_rating = 0
    for user in data:
        print(user)
        total_rating +=  float(user["rating"])
    print(f"avg rating is {total_rating/len(data)}")

    # percentage of users with poor rating
    poor_rating = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating += 1
    print(f"% of user with poor ratings = {poor_rating/len(data) * 100}%")
        
```

```python
get_insight(data)
```

```python
def recommendations(data):
    recommendations = []

    for user in data:
        current_recomm = {}
        current_recomm["name"] = user["name"]
        if(float(user["rating"]) >= 4):
            current_recomm["brand"] = "Apple"
        else:
             current_recomm["brand"] = "Samsung"
            
        recommendations.append(current_recomm)
    return recommendations
            
```

```python
recommendations(data)
```
