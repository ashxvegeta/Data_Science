```python
import numpy as np
import time
```

```python
size = 10_00_000

# python list
py_list = list(range(size))
start_time = time.time()
sq_list =  [x**2 for x in py_list]
end_time = time.time()
print(f"python list time = {end_time-start_time} seconds")

# numpy array
np_array = np.array(py_list)
start_time = time.time()
sq_array = np_array**2
end_time = time.time()
print(f"numpy array  time = {end_time-start_time} seconds")
```

```python
import sys

print(f"python list size = {sys.getsizeof(py_list)*len(py_list)} bytes")
print(f"numpy array size = {np_array.nbytes} bytes")
```
